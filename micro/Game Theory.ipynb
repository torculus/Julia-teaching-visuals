```julia
using PyPlot

x = 0:0.001:1

P1 = (1/2).*(sign.(x.-0.45).+1)

plot(x,P1, linestyle="--", linewidth=3.0)
xlabel("Player 1 strategy probability (σ)")
ylabel("Player 2 strategy probability (τ)")

P2 = (1/2).*(sign.(x.-0.3).+1)

plot(P2,x, linewidth=3.0)
```

```julia
using LightGraphs
using GraphPlot

g = SimpleGraph(7);

add_edge!(g, 1, 2);
add_edge!(g, 1, 3);
add_edge!(g, 2, 4);
add_edge!(g, 2, 5);
add_edge!(g, 3, 6);
add_edge!(g, 3, 7);

membership = [1,2,2,1,1,1,1]
nodecolor = ["lightseagreen", "orange"]
game_colors = nodecolor[membership]

game_label = ["P1","P2","P2","(8,3)","(6,4)","(8,6)","(0,1)"]
edge_label = ["A", "B", "C", "D", "C", "D"]

game_layout=(args...)->spring_layout(args...; C=4)

gplot(g, layout=game_layout, nodefillc=game_colors, nodelabel = game_label, edgelabel = edge_label)
```
