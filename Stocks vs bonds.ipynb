```julia
using PyPlot

x = 1:0.1:20

stock = 2 .*x.+ 19 .*(rand(length(x)) .- 0.5) .- 10sin.(x) .+ 1
stock_trend = 2 .* x .+ 1
bond = 1 .*x .+ 1
bank = 0.5 .*x .+ 1

plot(x, stock, color="green", label="stock");
plot(x, stock_trend, color="gray", linestyle="--");
plot(x, bond, color="blue", label="bond");
plot(x, bank, color="red", label="bank account");
xlabel("time")
ylabel("Value")
legend();
```

```julia
using PyPlot

x = 1:0.1:20

stock = 2 .*x.+ 19 .*(rand(length(x)) .- 0.5) .- 10sin.(x) .+ 1
stock_trend = 2 .* x .+ 1

sandp500 = 2 .*x.+ 3 .*(rand(length(x)) .- 0.5) .- 10sin.(x) .+ 1

plot(x, stock, color="green", label="1 stock");
plot(x, sandp500, color="blue", label="S&P 500 (average of 500 stocks)");
plot(x, stock_trend, color="gray", linestyle="--");
xlabel("time")
ylabel("Price")
legend();
```

```julia
using PyPlot

x = 1:1:65

y = [1.5 .*ones(10,1); 2 .*ones(10,1); 2.5 .* ones(15,1); 2 .*ones(10); 3 .* ones(20, 1)]

z = x./40 + 3 .*rand(length(x),1)

plot(x, y, label="sticky prices")
plot(x, z, color="green", label="flexible prices (Supply=Demand)")
xlabel("time")
ylabel("Price")
legend()
```

```julia
# calculate the Value function for all values of capital
K = 0:10:1000; # K is the space for both k and k'
# make an initial guess for V()
V0 = zeros(length(K));
beta = 0.9;
iter = 0;
diff = 1;
tol = 1.0e-4;
maxiter = 1e5;
while diff > tol
    iter += 1;
    V1 = K + beta*V0;
    diff = (V1-V0);
    V0 = V1;
    if iter > maxiter
        break
    end
end
```
