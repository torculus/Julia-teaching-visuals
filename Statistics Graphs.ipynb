```julia
using PyPlot

x = linspace(0,5,1000);
y = 3 .* x.^2 .* exp(-x.^3);
plot(x, y, color="green", linewidth=2.0, linestyle="-")
```

```julia
x = linspace(-3, 3, 1000)

z = 1/sqrt(2*pi) * exp.(-1*x.^2/2);
w = 1/sqrt(pi) * exp.(-2*x.^2);
v = 1/sqrt(pi/5) * exp.(-100 * x.^2);

plot(x, z, color="blue", linewidth=2.0, linestyle="-")
plot(x, w, color="green", linewidth=2.0, linestyle="-")
plot(x, v, color="black", linewidth=2.0, linestyle="-")
```

```julia
x = linspace(-3, 3, 1000)

n = 15

for i=1:n
    sigma = 1/i^2;
    y = 1/sqrt(2*pi*sigma) * exp.(-1*x.^2/ (sigma^2) );
    plot(x, y, color=(0.5, 0.5, i/n), linewidth=2.0, linestyle="-")
end
```

```julia
using PyPlot

x = linspace(0,10,100);
y = 2 + 3*x + 3*randn(length(x),1);
z = 2 + 3*x;
plot(x, y, color="green", linewidth=2.0, linestyle="-")
plot(x, z, color="black", linewidth=2.0, linestyle="--")
title("Stock price with trend")
grid("on")
```
