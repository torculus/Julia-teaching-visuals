```julia
using Plots, Interact

timer = Observable(1)

@sync while true
    sleep(1)
    timer[] = timer[]+1
end

@manipulate for s = slider(1:10), lw=1:0.1:5, t=timer
    x = -3:0.1:3
    plot(x, sin.(x.+t), linewidth=lw)
end
```

```julia
using Plots, Interact



x = linspace(-3, 3, 1000)

n = 15

for i=1:n
    sigma = 1/i^2;
    y = 1/sqrt(2*pi*sigma) * exp.(-1*x.^2/ (sigma^2) );
    plot(x, y, color=(0.5, 0.5, i/n), linewidth=2.0, linestyle="-")
end
```
