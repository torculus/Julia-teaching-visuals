```julia
using Plots;

x=0:40;
supply(x) = 3 .* x
demand(x) = 100 .- 2x;

@gif for t=0:25
    reces = plot(label="", xlims=(0,40), ylims=(0,100), title="Recession")
    plot!(reces, [20], seriestype=:vline, label="", lw=3, linecolor=:red)
    plot!(reces, x, supply(x)*(t<=20) + supply(x.-t/3)*(t>20), label="", lw=3, linecolor=:deeppink)
    plot!(reces, x, demand(x.+t)*(t<=10) + demand(x.+12)*(t>10), label="", lw=3, linecolor=:blue)
end every 1
```

```julia
using Plots;

x=0:40;
supply(x) = 3 .* x
demand(x) = 100 .- 2x;

@gif for t=0:25
    boom = plot(label="", xlims=(0,40), ylims=(0,100), title="Boom")
    plot!(boom, [20], seriestype=:vline, label="", lw=3, linecolor=:red)
    plot!(boom, x, supply(x)*(t<=20) + supply(x.+t/3)*(t>20), label="", lw=3, linecolor=:deeppink)
    plot!(boom, x, demand(x.-t)*(t<=10) + demand(x.-12)*(t>10), label="", lw=3, linecolor=:blue)
end every 1
```

```julia
using Plots;

x=0:40;
supply(x) = 2.5 .* x
demand(x) = 100 .- 2.5*x;

@gif for t=0:80
    # (0,10) demand, (10,20) rest, (20,30) supply, (30,40) rest, (40,50) demand, (50,60) rest, (60,70) supply, (70,80)
    busi = plot(label="", xlims=(0,40), ylims=(0,100), title="Business Cycle")
    plot!(busi, [20], seriestype=:vline, label="", lw=3, linecolor=:red)
    plot!(busi, x, demand(x.+t)*(t<=10) + demand(x.+12)*(10<t<=40), label="", lw=3, linecolor=:blue) #recession
    plot!(busi, x, supply(x)*(t<=20) + supply(x.-(t-20))*(20<t<=30), label="", lw=3, linecolor=:deeppink) #recession
    plot!(busi, x, demand(x.+12 .-(t-40))*(40<t<=50) + demand(x)*(t>50), label="", lw=3, linecolor=:blue) #boom
    plot!(busi, x, supply(x.-12)*(30<t<=60) + supply(x.-12 .+(t-60))*(60<t<=70) + supply(x)*(t>70), label="", lw=3, linecolor=:deeppink) #boom
end every 1
```
